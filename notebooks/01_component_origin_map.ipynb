# Notebook 01 — Country of origin of foreign components inside Russian drones

Reproduces the headline finding from the [KSE Institute analysis](https://kse.ua/about-the-school/news/foreign-components-in-russian-military-drones/):
**~69 % of foreign components inside Russian drones originate from US-owned companies**.

Inputs: `data/components.csv` (live from GUR War & Sanctions portal).

```python
import pandas as pd, matplotlib.pyplot as plt
df = pd.read_csv('../data/components.csv')
print(f'Loaded {len(df):,} component rows')
df.head()
```

```python
counts = df['manufacturer_country'].value_counts().drop('UNK', errors='ignore')
print(counts.head(15))
```

```python
fig, ax = plt.subplots(figsize=(9,5))
top10 = counts.head(10)
ax.barh(top10.index[::-1], top10.values[::-1])
ax.set_xlabel('Number of identified components')
ax.set_title('Top-10 countries of origin — components inside documented drones (GUR data)')
for i, v in enumerate(top10.values[::-1]):
    ax.text(v, i, f' {v:,}', va='center')
plt.tight_layout()
plt.savefig('../images/origin_countries_top10.png', dpi=150)
plt.show()
```

```python
us_share = counts.get('US', 0) / counts.sum() * 100
print(f'US-origin share of identified components: {us_share:.1f}%')
print('Compare with KSE Institute 2023 finding: 69%')
```

```python
# Top manufacturers
top_mfr = df[df['manufacturer'] != 'Not identified']['manufacturer'].value_counts().head(15)
fig, ax = plt.subplots(figsize=(9,6))
ax.barh(top_mfr.index[::-1], top_mfr.values[::-1])
ax.set_xlabel('Number of identified components')
ax.set_title('Top-15 manufacturers — components inside documented drones')
plt.tight_layout()
plt.savefig('../images/top_manufacturers.png', dpi=150)
plt.show()
```
